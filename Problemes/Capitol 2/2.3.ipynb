```python
def nota(num):
    """
    Aquesta funció retorna la qualificació d'un alumne.
    
    Parameters
    ----------
    num: float
       
    Returns
    -------
    qualificació: string
    """

    qualificacio = ""

    if num < 5:
        qualificacio = 'Suspès'
    elif num < 6:
        qualificacio = 'Aprovat'
    elif num < 8:
        qualificacio = 'Notable'
    elif num <10:
        qualificacio = 'Excel·lent'
    else:
        qualificacio = 'Matrícula d\'honor'

    return qualificacio
```

```python
assert nota(0.9) == 'Suspès'
assert nota(5.9) == 'Aprovat'
assert nota(7.9) == 'Notable'
assert nota(9.9) == 'Excel·lent'
assert nota(10) == 'Matrícula d\'honor'
```
