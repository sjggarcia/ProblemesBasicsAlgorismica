### 2.15 Nombres feliç

Un nombre és feliç si en transformar-lo de forma iterativa en la suma dels quadrats dels seus dígits, fins que només quedi una xifra, acabem amb un 1. Per exemple 203:
+ 203: $2^2 + 0^2 + 3^2 = 13$
+ 13: $1^2 + 3^2 = 10$
+ 10: $1^2 + 0^2 = 1$

Escriu una funció que determini si un nombre és feliç.

```python
def felic(num):
    """
    Aquesta funció comprova si un nombre és feliç.
    
    Parameters
    ----------
    num: int
        El número
    
    Returns
    -------
    b: bool
        Si el número és feliç    
    """

    b = False

    while num >= 10:
        suma = 0
        for xifra in str(num):
            suma += int(xifra)**2
        num = suma
    
    if num == 1:
        b = True
        
    return b

felic(203)
```

```python
assert felic(203) == True
assert felic(13) == True
assert felic(10) == True
assert felic(96) == False
assert felic(33) == False

assert felic(921) == True
assert felic(717) == False
assert felic(2039) == True
assert felic(8487) == True
assert felic(1) == True
```
