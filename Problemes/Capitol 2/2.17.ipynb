```python
def avet(num):
    """
    Aquesta funció dibuixa un avet.
    
    Parameters
    ----------
    num: int
        Nombre d'asterics a la part inferior
    
    Returns
    -------
    None
    """

    linia = "*"
    n = 1

    while n < num:
        linia *= n
        n += 2
        print(linia.strip())

    
avet(7)
```
