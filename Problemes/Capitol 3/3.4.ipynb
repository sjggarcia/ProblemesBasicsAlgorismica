```python
def resta_binaria(op1, op2):  
    """
    Aquesta funció calcula la resta binaria donat dos nombres.
    
    parameters
    ----------
    op1: llista d'enters
    op2: llista d'enters
    
    Returns
    -------
    resultat: llista d'enters
    """

    result = []
    n1 = 0
    n2 = 0
    potencia = 0

    for i in op1[::-1]:
      n1 += i*(2**potencia)
      potencia += 1
    
    potencia = 0
    for j in op2[::-1]:
      n2 += j*(2**potencia)
      potencia += 1

    resta = n1 - n2
    potencia = 0

    for k in range(1, resta):
        if k%(2**potencia) == 0:
            result.append(0)
        else:
           result.append(1)
        potencia += 1

    # el número en forma de lista tiene que tener 4 items
    result.append(0)

    return result[::-1]

resta_binaria([0,1,0,1], [0,0,0,1])
```
