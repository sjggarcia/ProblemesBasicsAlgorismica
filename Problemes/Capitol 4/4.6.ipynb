### 4.6 Run Length Encoding

```python
def rle(text):
    """
    Aquesta funció retorna un text codificat segons run length encoding.
    
    Parameters
    ----------
    text: string
        text a codificar
        
    Returns
    -------
    text: string
        text codificat
    """

    textcodificat = ""
    vegades = 1

    for i in range(len(text)-1):
        if text[i] != text[i+1]:
            textcodificat += text[i] + str(vegades)
            vegades = 1
        else:
            vegades += 1
            
    # es necessita fer això perquè el for se salta l'últim element 
    textcodificat += (text[-1] + str(vegades))
    return textcodificat

rle("ABBBBNNNEEEDDDZZAAAAA")
```

```python
assert rle("ABBBBNNNEEEDDDZZAAAAA") == 'A1B4N3E3D3Z2A5'
assert rle("BBBBBBBBBBBBBBBBBWWWWWWZAAA") == 'B17W6Z1A3'
```

       
La complexitat de l'algorisme és O(n).
