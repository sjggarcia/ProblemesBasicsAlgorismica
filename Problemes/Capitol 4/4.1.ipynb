### 4.1 Acronims

```python
def acronim(frase):
    """
    Aquesta funció retorna l'acrònim d'una frase.
    
    Parameters
    ----------
    frase: string
    
    Returns
    -------
    acronim: string
    """
    
    acronim = ""
    frase = frase.upper()
    frase = frase.split()

    for paraula in frase:
        acronim += paraula[0]

    return acronim

acronim("Hola que tal")
```

```python
assert acronim("Hola que tal") == 'HQT'
```
